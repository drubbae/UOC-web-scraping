```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

debug = False

url='https://cryptoslate.com/coins/'

#Create a handle, page, to handle the contents of the website
page = requests.get(url).text

soup = BeautifulSoup(page,'lxml')

rows_coins = soup.find('p',{'id':'lower-coin-count'}).get_text().replace('Displaying ','').replace(' total coins','').replace(' of ','-')
rows_coins_min = int(rows_coins[:rows_coins.find('-')])
if debug: print(str(rows_coins_min))
rows_coins = rows_coins[1+rows_coins.find('-'):]
rows_coins_max = int(rows_coins[:rows_coins.find('-')])
if debug: print(str(rows_coins_max))
rows_coins_total = int(rows_coins[1+rows_coins.find('-'):])
if debug: print(str(rows_coins_total))

My_table = soup.find('table',{'class':'objects-table'})

rows = My_table.findAll('th')
titles=[]
for row in rows:
    titles.append(row.get_text())
if debug: print(titles)

df = pd.DataFrame()

i = 1
values1=[]
values2=[]
values3=[]
values4=[]
values5=[]
values6=[]
values7=[]
while rows_coins_max < rows_coins_total:
    rows = My_table.findAll('td',{'class':'col name'})
    for row in rows:
        values1.append(row.get_text())
    if debug: print(values1)

    rows = My_table.findAll('td',{'class':'col market-cap'})
    for row in rows:
        values2.append(row.get_text().replace('$',''))
    if debug: print(values2)

    rows = My_table.findAll('td',{'class':'col price'})
    for row in rows:
        if (len(row.get_text()) > 0) and (row.get_text().replace('$','').replace(',','').isnumeric()):
            values3.append(float(row.get_text().replace('$','').replace(',','')))
        else:
            values3.append(0)
    if debug: print(values3)

    rows = My_table.findAll('td',{'class':'col supply'})
    for row in rows:
        values4.append(row.get_text().replace('N/A','0').replace(',',''))
    if debug: print(values4)

    rows = My_table.findAll('td',{'class':'col blockchain'})
    for row in rows:
        values5.append(row.get_text())
    if debug: print(values5)

    rows = My_table.findAll('td',{'class':'col change-24h'})
    for row in rows:
        if (len(row.get_text()) > 0) and (row.get_text().replace('N/A','0').replace('%','').replace(',','').isnumeric()):
            values6.append(float(row.get_text().replace('N/A','0').replace('%','').replace(',','')))
        else:
            values6.append(0)
    if debug: print(values6)

    rows = My_table.findAll('td',{'class':'col change-7d'})
    for row in rows:
        if (len(row.get_text()) > 0) and (row.get_text().replace('N/A','0').replace('%','').replace(',','').isnumeric()):
            values7.append(float(row.get_text().replace('N/A','0').replace('%','').replace(',','')))
        else:
            values7.append(0)
    if debug: print(values7)
    
    i += 1
    if debug: print("i = " + str(i))
    url='https://cryptoslate.com/coins/page/'+str(i)
    #Create a handle, page, to handle the contents of the website
    page = requests.get(url).text
    soup = BeautifulSoup(page,'lxml')
    
    rows_coins = soup.find('p',{'id':'lower-coin-count'}).get_text().replace('Displaying ','').replace(' total coins','').replace(' of ','-')
    rows_coins_min = int(rows_coins[:rows_coins.find('-')])
    if debug: print(str(rows_coins_min))
    rows_coins = rows_coins[1+rows_coins.find('-'):]
    rows_coins_max = int(rows_coins[:rows_coins.find('-')])
    if debug: print(str(rows_coins_max))
    rows_coins_total = int(rows_coins[1+rows_coins.find('-'):])
    if debug: print(str(rows_coins_total))

    My_table = soup.find('table',{'class':'objects-table'})

df[titles[2]] = values1
df[titles[4]] = values2
df[titles[5]] = values3
df[titles[9]] = values4
df[titles[10]] = values5
df[titles[11]] = values6
df[titles[12]] = values7
    
df.to_csv('crypto_currencies.csv')
```
